```python
import matplotlib.pyplot as plt
import random
```

```python
_x = [0]
_y = [0]

def my_plt(x, y, n):
    if n == 0:
      _x.append(x/2.0)
      _y.append(y/2.0)
    
    elif n == 1:
      _x.append((x + 1)/2.0)
      _y.append(y/2.0)

    else:
      _x.append((x + 0.5)/2.0)
      _y.append((y + (3**0.5))/2.0)
```

```python
for t in range(100000):
    n = random.randint(0, 2)
    my_plt(_x[-1], _y[-1], n)
```

```python
fig, ax = plt.subplots()
ax.scatter(_x, _y, color='black', s=0.0001)
ax.set_title('узор Серпинского')

fig.set_figwidth(13)    
fig.set_figheight(15)    
```
